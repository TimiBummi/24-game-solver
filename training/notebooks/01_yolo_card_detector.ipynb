### Remap Classes

The dataset may have fine-grained labels (e.g. `ace_of_spades`, `2_of_hearts`, …). We collapse all 52 card classes into a single class `card` (class ID 0) since Stage 1 only needs to *locate* cards — Stage 2 will classify the rank/suit.

````python
import shutil

# Back up original data.yaml and label files before remapping.
# Notebook 3 needs the original class IDs (0–51) to map each annotation to a rank.
# Both are backed up here, before any modification.

# 1. data.yaml → data_original.yaml  (preserves class name → index mapping)
orig_yaml = os.path.join(DATA_DIR, "data.yaml")
backup_yaml = os.path.join(DATA_DIR, "data_original.yaml")
if os.path.exists(orig_yaml) and not os.path.exists(backup_yaml):
    shutil.copy(orig_yaml, backup_yaml)
    print(f"Backed up data.yaml → data_original.yaml")

# 2. labels/ → labels_original/  (preserves original class IDs in .txt files)
for split in ["train", "test", "valid"]:
    src = os.path.join(DATA_DIR, split, "labels")
    dst = os.path.join(DATA_DIR, split, "labels_original")
    if os.path.isdir(src) and not os.path.isdir(dst):
        shutil.copytree(src, dst)
        print(f"Backed up {split}/labels/ → {split}/labels_original/")

# ── Remap all class labels to single class 0 ("card") ────────────────────────
# YOLO format: class_id center_x center_y width height (normalized)

def remap_labels_to_single_class(labels_dir: str) -> int:
    """Rewrite all YOLO label files to use class 0 for every object."""
    count = 0
    for label_file in glob.glob(os.path.join(labels_dir, "*.txt")):
        with open(label_file, "r") as f:
            lines = f.readlines()

        remapped = []
        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 5:
                parts[0] = "0"
                remapped.append(" ".join(parts))

        with open(label_file, "w") as f:
            f.write("\n".join(remapped) + "\n" if remapped else "")
        count += 1
    return count

for split in ["train", "test", "valid"]:
    labels_dir = os.path.join(DATA_DIR, split, "labels")
    if os.path.isdir(labels_dir):
        n = remap_labels_to_single_class(labels_dir)
        print(f"Remapped {n} label files in {split}/")
    else:
        print(f"No labels dir found for {split}")
````

### Create data.yaml

Write the YOLO dataset config file that points to the train/val splits and defines the single `card` class.

````python
# Create YOLO data config pointing to the Kaggle dataset
data_yaml = f"""
path: /content/data
train: train/images
val: valid/images
test: test/images

nc: 1
names:
  0: card
"""

with open(os.path.join(DATA_DIR, "data.yaml"), "w") as f:
    f.write(data_yaml.strip())

print("Created data.yaml")
print(open(os.path.join(DATA_DIR, "data.yaml")).read())
````

## Training

Train **YOLOv8-nano** for up to 100 epochs with early stopping (patience=15). Training at 640×640 gives better feature quality before we export at 320×320 for inference.

Expected training time on a T4 GPU: ~20–40 minutes depending on dataset size.

````python
# Resume from checkpoint if available, otherwise start fresh
SAVE_DIR = os.path.join(DRIVE_OUTPUT, "yolo_card_detector")
checkpoint = os.path.join(SAVE_DIR, "weights", "last.pt")

if os.path.exists(checkpoint):
    print(f"Resuming from {checkpoint}")
    model = YOLO(checkpoint)
    results = model.train(resume=True)
else:
    print("Starting fresh from pretrained weights")
    model = YOLO("yolov8n.pt")
    results = model.train(
        data="data/data.yaml",
        epochs=100,
        imgsz=640,
        batch=16,
        patience=15,
        project=DRIVE_OUTPUT,
        name="yolo_card_detector",
        exist_ok=True,
    )

BEST_PT = f"{SAVE_DIR}/weights/best.pt"

print("Training complete!")
print(f"Best model: {BEST_PT}")
assert os.path.exists(BEST_PT), f"best.pt not found at {BEST_PT}"
````

## Evaluation

Evaluate the best checkpoint on the validation set. We target **mAP@0.5 > 0.90** for reliable card detection in the Flutter app.

````python
# Evaluate on validation set
best_model = YOLO(BEST_PT)
metrics = best_model.val(data="data/data.yaml")

print(f"mAP@0.5:     {metrics.box.map50:.4f}")
print(f"mAP@0.5:0.95: {metrics.box.map:.4f}")
print(f"Precision:    {metrics.box.mp:.4f}")
print(f"Recall:       {metrics.box.mr:.4f}")

# Target: mAP@0.5 > 0.90
assert metrics.box.map50 > 0.85, f"mAP@0.5 too low: {metrics.box.map50:.4f}"
````

### Visualize Predictions

Run inference on a sample of validation images and display the annotated results.

````python
import matplotlib.pyplot as plt
from PIL import Image

# Run predictions on sample validation images
val_images = glob.glob("data/valid/images/*.jpg")[:6]
if not val_images:
    val_images = glob.glob("data/valid/images/*.png")[:6]

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
for ax, img_path in zip(axes.flat, val_images):
    results = best_model.predict(img_path, conf=0.5, verbose=False)
    annotated = results[0].plot()
    ax.imshow(annotated[:, :, ::-1])  # BGR to RGB
    ax.axis("off")
    ax.set_title(Path(img_path).name)

plt.suptitle("YOLOv8-nano Card Detector — Validation Predictions", fontsize=16)
plt.tight_layout()
os.makedirs("outputs/yolo_card_detector", exist_ok=True)
plt.savefig("outputs/yolo_card_detector/val_predictions.png", dpi=150)
plt.show()
````

## Export

Export the trained model to **TFLite FP16** at 320×320 resolution for on-device inference.

After exporting, copy the `.tflite` file to `flutter_app/assets/models/card_detector.tflite`.

````python
# Export to TFLite at inference size 320x320
# FP16 quantization for good accuracy + small size
exported_path = best_model.export(
    format="tflite",
    imgsz=320,
    int8=False,
    half=True,  # FP16
)

tflite_files = glob.glob(os.path.join(str(SAVE_DIR), "weights", "*float16.tflite")) + \
               glob.glob(os.path.join(str(SAVE_DIR), "weights", "*.tflite"))
for f in tflite_files:
    size_mb = os.path.getsize(f) / (1024 * 1024)
    print(f"{f}: {size_mb:.1f} MB")

print("\nCopy the .tflite file to flutter_app/assets/models/card_detector.tflite")
````

## Real Photo Evaluation

Test the trained detector on your own photos to measure the domain gap between synthetic training data and real-world images.

**Upload your `real_photos/` folder** to Colab (drag into the file browser, or use the cell below to upload from Google Drive).

````python
# Upload real photos — pick one method:

# Option A: Upload folder manually via Colab file browser (drag & drop to /content/real_photos/)

# Option B: Upload from Google Drive
# from google.colab import drive
# drive.mount('/content/drive')
# !cp -r "/content/drive/MyDrive/real_photos" /content/real_photos

import json

REAL_PHOTOS_DIR = "real_photos"
GT_PATH = os.path.join(REAL_PHOTOS_DIR, "ground_truth.json")

real_images = sorted(glob.glob(os.path.join(REAL_PHOTOS_DIR, "*.jpg")))
print(f"Found {len(real_images)} real photos")

# Load ground truth
with open(GT_PATH) as f:
    ground_truth = json.load(f)
# Remove metadata key
ground_truth.pop("_note", None)
print(f"Ground truth entries: {len(ground_truth)}")
````

````python
# Run detector on all real photos and compare detected card count vs ground truth
correct_count = 0
total = 0
results_log = []

for img_path in real_images:
    stem = Path(img_path).stem
    if stem not in ground_truth:
        continue

    gt = ground_truth[stem]
    expected_cards = gt["cards"]

    results = best_model.predict(img_path, conf=0.5, verbose=False)
    detected_cards = len(results[0].boxes)

    match = detected_cards == expected_cards
    if match:
        correct_count += 1
    total += 1

    results_log.append({
        "image": stem,
        "expected": expected_cards,
        "detected": detected_cards,
        "match": match,
    })

    if not match:
        print(f"  MISMATCH {stem}: expected {expected_cards}, detected {detected_cards}")

print(f"\nCard count accuracy: {correct_count}/{total} ({correct_count/total:.1%})")
print(f"Images with wrong count: {total - correct_count}")
````

````python
# Visualize predictions on a sample of real photos
sample = real_images[:6]
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
for ax, img_path in zip(axes.flat, sample):
    results = best_model.predict(img_path, conf=0.5, verbose=False)
    annotated = results[0].plot()
    ax.imshow(annotated[:, :, ::-1])
    ax.axis("off")
    stem = Path(img_path).stem
    gt_count = ground_truth.get(stem, {}).get("cards", "?")
    det_count = len(results[0].boxes)
    ax.set_title(f"{stem} — GT: {gt_count}, Det: {det_count}", fontsize=11)

plt.suptitle("YOLOv8-nano — Real Photo Predictions", fontsize=16)
plt.tight_layout()
plt.savefig("outputs/yolo_card_detector/real_photo_predictions.png", dpi=150)
plt.show()
````
